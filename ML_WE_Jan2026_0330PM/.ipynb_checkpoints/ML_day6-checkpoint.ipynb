```python
# PCA (Principal Component Analysis)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.shape
```

```python
df.head()
```

```python
newdf = df.copy()
```

```python
df = df.drop(columns=['species'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.decomposition import PCA
```

```python
scaled_df = StandardScaler().fit_transform(df)
```

```python
scaled_df[:4]
```

```python
scaled_df = pd.DataFrame(scaled_df , columns=df.columns)
```

```python
scaled_df.head(3)
```

```python
scaled_df.describe()
```

```python
pca = PCA( n_components = 2 )
```

```python
scaled_df = pca.fit_transform(scaled_df)
```

```python
scaled_df[:4]
```

```python
scaled_df = pd.DataFrame(scaled_df , columns=['pca1', 'pca2'])
```

```python
scaled_df.head(3)
```

```python
scaled_df['species'] = newdf['species']
```

```python
scaled_df.head(3)
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
sns.scatterplot( x='sepal_length',y='sepal_width' ,hue='species' , data=newdf )
plt.subplot(1,2,2)
sns.scatterplot( x='pca1',y='pca2' ,hue='species', data=scaled_df)
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
sns.distplot(scaled_df['pca1'])
plt.subplot(1,2,2)
sns.distplot(scaled_df['pca2'])
plt.show()
```

```python
df = scaled_df.drop(columns=['species'])
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans( n_clusters=3 )
```

```python
cluster = kmeans.fit_predict(df)
```

```python
cluster
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
sns.scatterplot(x='pca1',y='pca2',hue='species',data=scaled_df)
plt.subplot(1,2,2)
sns.scatterplot(x='pca1',y='pca2',hue=cluster,data=df)
plt.subplot(1,2,2)
plt.show()
```
