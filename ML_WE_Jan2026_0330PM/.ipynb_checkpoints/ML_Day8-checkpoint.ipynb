```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
```

```python
df.shape
```

```python
df.head(2)
```

```python
df = df.drop(columns=['deck','alive','alone','adult_male','class','embark_town'])
```

```python
df.shape
```

```python
df.head(2)
```

```python
for col in df.select_dtypes(include='object').columns:
    print(col,'\t',df[col].unique())
```

```python
# filling missing values with mode in all categorical columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(value=df[col].mode()[0]) 
```

```python
for col in df.select_dtypes(include='object').columns:
    print(col,'\t',df[col].unique())
```

```python
df.info()
```

```python
# Applying LabelEncoding
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].map( lambda val: list(df[col].unique()).index(val) )
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
# filling missing value in float columns using average
for col in df.select_dtypes(include='float').columns:
    df[col] = df[col].fillna(value = round(df[col].mean(),2))
```

```python
df.isnull().sum()
```

```python
for col in df.select_dtypes(include='float').columns:
    print(col)
```

```python
plt.figure(figsize=(10,3))
i = 1
for col in df.select_dtypes(include='float').columns:
    plt.subplot(1,2,i)
    sns.boxplot(df[col])
    plt.title(col)
    i = i+1
plt.show()
```

```python
# Applying MinMaxScaler for normalization (converting all data from 0 to 1)
# ( X-X.mean() )/(Xmax - Xmin)
for col in df.select_dtypes(include='float').columns:
    df[col] = ( df[col]-df[col].min() )/( df[col].max() - df[col].min() )
```

```python
df.head()
```

```python
plt.figure(figsize=(10,3))
i = 1
for col in df.select_dtypes(include='float').columns:
    plt.subplot(1,2,i)
    sns.boxplot(df[col])
    plt.title(col)
    i = i+1
plt.show()
```

```python
X = df.drop(columns=['survived'])
y = df['survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split( X,y, random_state=11 , test_size=0.2 )
```

```python
X_train.shape , X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
```

```python
print( 'Accuracy :',accuracy_score( y_pred,y_test )*100 )
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print( classification_report( y_pred , y_test ) )
```
