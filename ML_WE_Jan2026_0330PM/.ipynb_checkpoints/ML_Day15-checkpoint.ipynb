### Movie Recommendation System

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = {
    'movie_id':[x for x in range(101,109)] , 
    'movie_name':['Action Hero','Love Forever','Space Wars','Romantic Escape',
                  'Comedy Nights','Action Reloaded','Galaxy Battle','Laugh Out Loud'],
    'movie_genre':['Action','Romance','Sci-Fi','Romance','Comedy','Action','Sci-Fi','Comedy']
}
```

```python
df = pd.DataFrame(data)
df.shape
```

```python
vectorizer = CountVectorizer()
```

```python
genre_matrix = vectorizer.fit_transform(df['movie_genre'])
genre_matrix.toarray()
```

```python
similarity = cosine_similarity(genre_matrix)
similarity
```

```python
# Movie Recommendation Method
def recommend_movie(title):
    index = df[df['movie_name']==title].index[0]
    scores = list(enumerate(similarity[index]))
    sorted_scores = sorted(scores , key=lambda x:x[1] , reverse=True)
    sorted_scores = sorted_scores[1:3]   # top 2
    recom = []
    for i in sorted_scores:
        recom.append( df.iloc[i[0]]['movie_name'] )
    return recom
```

```python
recommend_movie('Space Wars')
```
