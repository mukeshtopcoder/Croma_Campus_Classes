```python
for col in df.columns:
    if df[col].isnull().sum()>0:
        print(col,"\t",df[col].dtype,end="")
        if df[col].dtype=='O':
            print("\t\tMissing Replaced By",df[col].mode()[0])
            df[col]= df[col].fillna(value=df[col].mode()[0])
        else:
            print("\tMissing Replaced By",round(df[col].mean(),2))
            df[col]= df[col].fillna(value=round(df[col].mean(),2))
```

```python
df['z_score'] = (df['Profit']-df['Profit'].mean())/df['Profit'].std()
```

```python
df['z_score']
```

```python
df = df[df['z_score']>-3]
df = df[df['z_score']< 3]
df = df.drop(columns=['z_score'])
```
