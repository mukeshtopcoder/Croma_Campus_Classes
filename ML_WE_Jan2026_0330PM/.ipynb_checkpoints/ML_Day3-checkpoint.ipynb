```python
import pandas as pd
import numpy as np
```

```python
years = [1,2,3,4,5,6,7]
salary = [25,30,35,40,45,50,55]
```

```python
# Regression
# y = mx+c
# X => input (Years)
# Y => Output (Salary)
# m (slope) , c (intercept) 
```

```python
# enter input x => y=mx+c => y(output)
```

```python
m = 5
c = 20
x = 9
y = m*x+c
y
```

```python
# Ex
Ex = sum( years )
# Ey
Ey = sum( salary )
# Exy
xy = [a*b for a,b in zip(years,salary)]
Exy = sum(xy)
# Ex2
x2 = [a**2 for a in years]
Ex2 = sum(x2)
# n
n = len(years)
# ymean 
ym = sum(salary)/len(salary)
# xmean
xm = sum(years)/len(years)
```

```python
m = ( n*Exy -Ex*Ey )/(n*Ex2 - Ex**2)
m
```

```python
c = ym - m*xm
c
```

```python
x = int(input("How Many Years Of Experience : "))
print("Predicted Salary :",(m*x+c))
```
