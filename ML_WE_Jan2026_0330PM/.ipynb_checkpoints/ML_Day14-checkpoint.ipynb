### Association Rules in Machine Learning

```python
# Bread , Milk , Jam
# Bread , Milk , Butter
# Bread , Butter
# Milk , Butter
# Bread , Jam , Milk
# Milk , Butter , Jam , Bread
```

#### 1. Itemset
A collection of items purchased together<br>
Example Transactions<br>
- Bread , Milk , Jam
- Bread , Milk , Butter
- Bread , Butter
- Milk , Butter
- Bread , Jam , Milk
- Milk , Butter , Jam , Bread

#### 2- Support
support = Total no transaction (M & B) / total transaction  =>  4/6 => 0.66

#### 3- Confidence
confidence = Support(M & B)/Support(M) => 4/5 => 0.80

#### 4- Lift
- lift = 1 , No Association
- lift < 1 , Negative Association
- lift > 1 , Positive Association <br>
lift = Confidence (A->B)/Support(A&B) => 0.88/0.66 => 1.33

```python
data = [
    ['Bread' , 'Milk' , 'Jam'],
    ['Bread' , 'Milk' , 'Butter'],
    ['Bread' , 'Butter'],
    ['Milk' , 'Butter'],
    ['Bread' , 'Jam' , 'Milk'],
    ['Milk' , 'Butter' , 'Jam' , 'Bread']
]
```

```python
# !pip install mlxtend
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori , association_rules
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
te_data = te.fit(data).transform(data)
```

```python
te_data
```

```python
df = pd.DataFrame(te_data , columns=te.columns_)
```

```python
df
```

```python
frequent_itemsets = apriori(df , min_support=0.4 , use_colnames = True)
```

```python
frequent_itemsets
```

```python
rules = association_rules( frequent_itemsets , metric='confidence' )
```

```python
rules[['antecedents','consequents','support','confidence','lift']]
```
