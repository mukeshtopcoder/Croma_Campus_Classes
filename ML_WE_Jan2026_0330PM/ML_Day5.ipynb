```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.shape
```

```python
df.head()
```

```python
df.species.unique()
```

```python
len(df.species.unique())
```

```python
df.head()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x='sepal_length',y='sepal_width',data=df)
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x='sepal_length',y='sepal_width',data=df,hue='species')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x='petal_length',y='petal_width',data=df,hue='species')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans( n_clusters=3 , random_state=42 )
```

```python
newdf = df.drop(columns=['species'])
```

```python
newdf.head()
```

```python
clusters = kmeans.fit_predict(newdf)
```

```python
clusters
```

```python
plt.figure(figsize=(6,3))
plt.scatter(newdf['petal_length'],newdf['petal_width'],c = clusters)
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.subplot( 1,2,1 )
sns.scatterplot(x='petal_length',y='petal_width',data=df,hue='species')
plt.subplot( 1,2,2 )
plt.scatter(newdf['petal_length'],newdf['petal_width'],c = clusters)
plt.show()
```

```python
clusters = kmeans.fit_predict( newdf[['petal_length','petal_width']] )
```

```python
clusters
```

```python
plt.figure(figsize=(10,3))
plt.subplot( 1,2,1 )
sns.scatterplot(x='petal_length',y='petal_width',data=df,hue='species')
plt.subplot( 1,2,2 )
plt.scatter(newdf['petal_length'],newdf['petal_width'],c = clusters)
plt.show()
```

```python
# AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering( n_clusters=3 )
```

```python
clusters = model.fit_predict(newdf)
```

```python
clusters
```

```python
plt.figure(figsize=(10,3))
plt.subplot( 1,2,1 )
sns.scatterplot(x='petal_length',y='petal_width',data=df,hue='species')
plt.subplot( 1,2,2 )
plt.scatter(newdf['petal_length'],newdf['petal_width'],c = clusters)
plt.show()
```

```python
clusters = model.fit_predict(newdf[['petal_width','petal_length']])
```

```python
clusters
```

```python
plt.figure(figsize=(10,3))
plt.subplot( 1,2,1 )
sns.scatterplot(x='petal_length',y='petal_width',data=df,hue='species')
plt.subplot( 1,2,2 )
plt.scatter(newdf['petal_length'],newdf['petal_width'],c = clusters)
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components=3 , random_state=42)
```

```python
clusters = gmm.fit_predict(newdf)
```

```python
clusters
```
