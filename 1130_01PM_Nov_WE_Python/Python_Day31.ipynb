```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\Croma Campus\Downloads\Financial_Sample.xlsx")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
plt.figure( figsize=(6,3) )
sns.heatmap(df.isnull())
plt.show()
```

```python
print("ColName\t\tAverage\t\tSTS\tStatus")
for col in df.select_dtypes(include=np.number).columns:
    print(col[:12],'\t\t',round(df[col].mean(),2),'\t\t',round(df[col].std(),2),end="")
    if df[col].mean()<df[col].std():
        print("\t\tOutlier")
    else:
        print("\t\tNot Outlier")
```

```python
i = 1
for col in df.select_dtypes(include=np.number).columns:
    plt.subplot(5,2,i)
    i=i+1
    sns.boxplot(df[col])
plt.show()
```
