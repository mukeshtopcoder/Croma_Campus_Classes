```python
# Step 2
# Load Libraries and Dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('titanic')
```

```python
# Step 4
# Data Cleaning
```

```python
# Replacing missing values in age column with mean/average
df['age'] = df['age'].fillna(df['age'].mean())

# Removing Deck column in dataset because it has too many missing values
df = df.drop(columns=['deck'])
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Step 5 
# Data Visualization
```

##### Survival Counts

```python
plt.figure( figsize=(6,3) )
sns.countplot( x='alive' , data=df )
plt.title("Survival Counts")
plt.show()
```

##### Survival By The Gender

```python
plt.figure( figsize=(6,3) )
sns.countplot( x='alive' , hue='sex' , data=df )
plt.title("Survival Counts By Gender")
plt.show()
```

##### Survival Cases by Passenger Class

```python
plt.figure( figsize=(6,3) )
sns.countplot( x='class' , hue='alive' , data=df )
plt.title("Survival Counts By Passenger class")
plt.show()
```

##### Age Distribution

```python
plt.figure(figsize=(6,3))
sns.histplot( df['age'] )
plt.title("Age Distribution")
plt.yticks( np.arange( 0, 250 , 25 ) )
plt.show()
```

##### Survival Cases By Age Distribution

```python
plt.figure(figsize=(6,3))
sns.histplot( x='age' , hue='alive' , data=df )
plt.title("Age Distribution")
plt.yticks( np.arange( 0, 250 , 25 ) )
plt.show()
```

##### Correlation Heatmap

```python
plt.figure(figsize=(5,3))
sns.heatmap(df.corr( numeric_only=True ) , annot=True )
plt.title("Correlation Matrix")
plt.show()
```

```python
# Step 6
# Key Insights
```

- Females had a higher survival rate than males
- First-class passengers survived more
- Yonger passengers had slightly better survival
- Fare price had some relation with survival cases
