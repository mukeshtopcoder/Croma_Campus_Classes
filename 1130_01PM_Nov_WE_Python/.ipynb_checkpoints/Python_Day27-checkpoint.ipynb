```python
def country(val):
    if val=='United States of America':
        return 'USA'
    else:
        return val[0:3].upper()
```

```python
country('United States of America')
```

```python
country('Germany')
```

```python
df.head(2)
```

```python
df.Country = df.Country.map(country)
```

```python
df['GrossManuAmt'] = df['Units Sold']*df['Manufacturing Price']
```

```python
df['Month Name'] = df['Month Name'].map(lambda val:val[0:3])
```

```python
for col in df.select_dtypes(include='float').drop(columns=['Units Sold']).columns:
    df[col] = df[col].map( lambda val:round(val/90.56, 2) )
```
