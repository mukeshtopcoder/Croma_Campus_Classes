```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(content.text , "html.parser")
```

```python
len(soup)
```

```python
soup.find('a',class_='no-underline flex-1 d-flex flex-column')
```

```python
topics = []
desc = []
link = []
```

```python
for row in soup.find_all('a',class_='no-underline flex-1 d-flex flex-column'):
    topics.append(row.find('p').text.strip())
    desc.append(row.find_all('p')[1].text.strip())
    link.append("https://github.com"+row['href'])
```

```python
data = {
    'Topic_Names':topics,
    'Link':link,
    'Description':desc
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
df.shape
```

```python
df.head(2)
```

```python
df['Link'][0]
```

```python
users = []
repo = []
link = []
star = []
```

```python
for url in df['Link']:
    content = requests.get(url)
    soup = BeautifulSoup(content.text , 'html.parser')
    for row in soup.find_all('div',class_='d-flex flex-justify-between flex-items-start flex-wrap gap-2 tmp-my-3'):
        users.append(row.find('a').text.strip())
        repo.append(row.find_all('a')[1].text.strip())
        link.append('https://github.com'+row.find_all('a')[1]['href'].strip())
        star.append(row.find('span',id="repo-stars-counter-star")['title'].strip())
```

```python
data = {
    'Username':users,
    'Repository':repo,
    'Rating':star,
    'Link':link
}
```

```python
df = pd.DataFrame(data)
```
