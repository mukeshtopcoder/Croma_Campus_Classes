```python
# When u have missing values in object columns => fill by dummy value or mode value
```

```python
df['Discount Band'].isnull().sum()
```

```python
df['Discount Band']
```

```python
# Fill NaN with Dummy Value
df['Discount Band'].fillna(value='Not Found')
```

```python
df['Discount Band'].mode()
```

```python
df['Discount Band'].mode()[0]
```

```python
df['Discount Band']
```

```python
df['Discount Band'].fillna(value=df['Discount Band'].mode()[0])
```

```python
df.columns
```

```python
df.select_dtypes(include='object').head(2)
```

```python
df.select_dtypes(include='object').columns
```

```python
for col in df.select_dtypes(include='object').columns:
    print(col)
```

```python
# Filling all missing values of object columns with MODE 
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(value= df[col].mode()[0] )
```

```python
df.isnull().sum()
```

```python
# fill missing data in float column =>  by average value
```

```python
df.columns
```

```python
df.select_dtypes(include='float').columns
```

```python
df['Units Sold'].mean()
```

```python
round(df['Units Sold'].mean())
```

```python
round(df['Units Sold'].mean() , 2)
```

```python
for col in df.select_dtypes(include='float').columns:
    print(col)
```

```python
# fill missing values by average in float columns
for col in df.select_dtypes(include='float').columns:
    df[col] = df[col].fillna(value=  round(df[col].mean(),2) )
```
