```python
sales = {
    'Month':['Jan','Feb','Mar','May','june'] , 
    'Sales':[87895,36693,57884,20974,98483]
}
```

```python
sales
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(sales)
```

```python
df
```

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar( df['Month'] , df['Sales']  )
```

```python
plt.figure( figsize=(6,3) )
plt.bar( df['Month'] , df['Sales'] )
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot( df['Month'] , df['Sales'] )
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.pie( df['Sales'] )
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.pie( df['Sales'] , labels=df['Month'])
plt.show()
```

```python
plt.figure( figsize=(10,4) )
plt.pie( df['Sales'] , labels=df['Sales'])
plt.legend(df['Month'])
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.pie( df['Sales'] , labels=df['Month'] , autopct="%.2f%%" )
plt.show() 
```
