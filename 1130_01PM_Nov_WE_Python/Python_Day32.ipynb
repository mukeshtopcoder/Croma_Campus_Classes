```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"C:\Users\Croma Campus\Downloads\Financial_Sample.xlsx").bfill()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
pbs = df.groupby("Segment").agg({'Profit':'sum'}).reset_index()
pbs
```

```python
plt.figure(figsize=(6,3))
sns.barplot( x="Segment" , y='Profit' , data=pbs )
plt.show()
```

```python
df = sns.load_dataset('tips')
```

```python
df.head(3)
```

```python
plt.figure(figsize=(6,3))
sns.barplot( x='day' , y='tip' , data=df )
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.violinplot( x='day' , y='tip' , data=df , palette='deep')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.violinplot( x=df['total_bill'])
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.violinplot( x=df['total_bill'] , color='salmon')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.violinplot( x=df['total_bill'] , color='teal')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.stripplot( x='day' , y='tip' , data=df , palette='deep')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.swarmplot( x='day' , y='tip' , data=df , palette='deep')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.pairplot(df , hue='day')
plt.grid()
plt.show()
```
