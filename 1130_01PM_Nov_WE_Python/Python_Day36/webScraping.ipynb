```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.boxofficemojo.com/year"
```

```python
# I want to scarp all data from this page and store it in a dataframe
# including the year link also
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")  
table = soup.find("table")  # find the first table in the page
rows = table.find_all("tr")  # find all rows in the table
data = []
```

```python
for row in rows:
    cols = row.find_all("td")  # find all columns in the row
    if len(cols) > 0:  # check if the row has columns
        year = cols[0].text.strip()  # get the year text
        link = cols[0].find("a")["href"]  # get the link from the first column
        data.append({"Year": year, "Link": "https://www.boxofficemojo.com" + link})  # append to data list

df = pd.DataFrame(data)  # create a DataFrame from the data list
```
