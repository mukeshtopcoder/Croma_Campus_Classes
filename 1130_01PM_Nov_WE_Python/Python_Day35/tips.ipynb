```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
# Loading the dataset
df = sns.load_dataset('tips')
```

```python
# show 2 rows of the dataset
df.head(2)
```

```python
# size of the dataset   
df.shape
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# distribution of total_bill column
plt.figure(figsize=(6, 3))  
sns.histplot(df['total_bill'], bins=30, kde=True)
plt.show()
```

```python
# Distribution of total_bill column with respect to the day
plt.figure(figsize=(6, 3))  
sns.histplot(data=df, x='total_bill', hue='day', bins=30, kde=True)
plt.show()
```

```python
# Distribution of tips column with respect to the sex
plt.figure(figsize=(6, 3))
sns.histplot(data=df, x='tip', hue='sex', bins=30, kde=True)
plt.show()
```
