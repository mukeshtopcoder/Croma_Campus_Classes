```python
for col in df.columns:
    if df[col].isnull().sum()>0:
        if df[col].dtype=='object':
            df[col] = df[col].fillna(value=df[col].mode()[0])
        else:
            df[col] = df[col].fillna(value=df[col].mean())
```
