```python
def convertDuration(val):
    li = val.split()
    if len(li)==2:
        return int(li[0][0:len(li[0])-1])*60+int(li[1][0:len(li[1])-1])
    elif li[0][-1]=='h':
        return int(li[0][0:len(li[0])-1])*60
    elif li[0][-1]=='m':
        return int(li[0][0:len(li[0])-1])
    else:
        return convertDuration(df.Duration.mode()[0])
```

```python
df.Duration = df.Duration.map(convertDuration)
```

```python
def convertVotes(val):
    val = str(val)
    if val[-1]=='K':
        return int(float(val[0:len(val)-1])*1000)
    elif val[-1]=='M':
        return int(float(val[0:len(val)-1])*1000000)
    else:
        return int(val)
```

```python
df.Votes = df.Votes.map(convertVotes)
```
