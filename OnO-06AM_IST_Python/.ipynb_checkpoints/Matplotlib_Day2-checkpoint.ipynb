```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Months':['Jan','Feb','Mar','Apr','May','Jun'],
    'Sales':[34545,56543,67865,23456,56565,23567]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.figure(figsize=(6,3))
plt.bar( df['Months'],df['Sales'] )
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar( df['Months'],df['Sales'] )
plt.xlabel("Months")
plt.ylabel("Sales Amount INR")
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar( df['Months'],df['Sales'] )
plt.xlabel("Months")
plt.ylabel("Sales Amount INR")
plt.title('Sales by Month')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar( df['Months'],df['Sales'] ,color='lightgreen')
plt.xlabel("Months")
plt.ylabel("Sales Amount INR")
plt.title('Sales by Month')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.barh( df['Months'],df['Sales'] ,color='lightgreen')
plt.ylabel("Months")
plt.xlabel("Sales Amount INR")
plt.title('Sales by Month')
plt.show()
```

```python
data = {
    'Months':['Jan','Feb','Mar','Apr','May','Jun'],
    'Sales':[34545,56543,67865,23456,-8700,23567]
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(6,3))
plt.bar( df['Months'],df['Sales'] ,color='lightgreen')
plt.xlabel("Months")
plt.ylabel("Sales Amount INR")
plt.title('Sales by Month')
plt.show()
```

```python
color = [ 'lightgreen' if val>0 else 'red' for val in df['Sales'] ]
```

```python
color
```

```python
plt.figure(figsize=(6,3))
plt.bar( df['Months'],df['Sales'] ,color=color)
plt.xlabel("Months")
plt.ylabel("Sales Amount INR")
plt.title('Sales by Month')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar( df['Months'],df['Sales'] ,color=color , width=0.4)
plt.xlabel("Months")
plt.ylabel("Sales Amount INR")
plt.title('Sales by Month')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar( df['Months'],df['Sales'] ,color=color , width=1)
plt.xlabel("Months")
plt.ylabel("Sales Amount INR")
plt.title('Sales by Month')
plt.show()
```

```python
data = {
    'Months':['Jan','Feb','Mar','Apr','May','Jun'],
    'Sales':[34545,56543,67865,23456,48700,23567]
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(6,3))
plt.pie(df['Sales'])
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(df['Sales'] ,labels=df['Months'])
# plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(df['Sales'] ,labels=df['Months'])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(df['Sales'] ,labels=df['Sales'])
plt.legend(df['Months'])
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(df['Sales'] ,labels=df['Months'])
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(df['Sales'] ,labels=df['Months'] , autopct="%.2f%%")
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(df['Sales'] ,labels=df['Months'] , autopct="%.2f%%")
cir = plt.Circle((0,0), 0.55 , fc='white')
plt.gca().add_artist(cir)
plt.show()
```
