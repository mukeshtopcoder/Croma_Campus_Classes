```python
x = ['Jan','Feb','Mar','Apr','May','Jun']
profit = [35560,56650,12340,45650,56760,23440]
sales = [75830,42360,84750,26330,48690,27530]
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,profit)
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,sales)
plt.plot(x,profit)
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,sales)
plt.plot(x,profit)
plt.grid()
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,sales)
plt.plot(x,profit)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,sales)
plt.plot(x,profit)
plt.grid(axis='x' , color='green' , ls=":")
plt.show()
```

```python
x = ['Jan','Feb','Mar','Apr','May','Jun']
profit = [35560,56650,12340,45650,56760,23440]
sales = [75830,42360,84750,26330,48690,27530]
```

```python
plt.figure(figsize=(6,3))
plt.subplot(2,1,1)
plt.plot(x,sales)
plt.subplot(2,1,2)
plt.plot(x,profit)
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.plot(x,sales)
plt.subplot(1,2,2)
plt.plot(x,profit)
plt.show()
```

```python
x1 = [23,78,85,123,67,90,78,45,23,45,67,90,45,67,89,34,54,89,56,34,67,89,34,67,89]
y1 = [536,786,485,623,687,940,768,485,345,678,123,474,567,257,565,584,849,256,384,637,789,834,667,889,836]
x2 = [ x+40 for x in x1 ]
y2 = [ y+90 for y in y1 ]
```

```python
plt.figure( figsize=(6,3) )
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.scatter(x1,y1 , color='red')
plt.scatter(x2,y2 , color='green')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.scatter(x1,y1  ,c=x1, cmap="viridis")
plt.scatter(x2,y2 , c=x2, cmap="plasma" )
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.scatter(x1,y1  ,c=x1, cmap="viridis")
plt.scatter(x2,y2 , c=x2, cmap="plasma" )
plt.colorbar()
plt.show()
```
