```python
# numerical/float64 replace null values by mean (average)
```

```python
df.Profit.mean()
```

```python
df.Segment = df.Segment.fillna(value='Government')
```

```python
df.Segment.isnull().sum()
```

```python
# Print only object Column Names
df.select_dtypes(include="object").columns
```

```python
# Fill all Object null value with mode
for col in df.select_dtypes(include="object").columns:
    if df[col].isnull().sum()>0:
        print(col,"\t\tNull Values : ",df[col].isnull().sum(),"\t\tMode :",df[col].mode()[0])
        df[col] = df[col].fillna(value=df[col].mode()[0])
```

```python
df.select_dtypes(include="object").isnull().sum()
```

```python
for col in df.select_dtypes(include="float64").columns:
    if df[col].isnull().sum()>0:
        print(col,"\t\tNull :",df[col].isnull().sum(),"\t\tAverage :",df[col].mean())
        df[col] = df[col].fillna(value=df[col].mean())
```

```python
df.select_dtypes(include="float64").isnull().sum()
```
