```python
months = [1,2,3,4,5,6]
sales = [234,787,345,896,456,798]
profit = [122,345,123,678,345,244]
```

```python
plt.figure(figsize=(6,3))
plt.plot(months,sales)
plt.plot(months,profit)
plt.xlabel('Months')
plt.ylabel("Amount in Lac")
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.plot(months,sales)
plt.plot(months,profit)
plt.title('Sales Analysis')
plt.xlabel('Months')
plt.ylabel("Amount in Lac")
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.plot(months,sales)
plt.plot(months,profit)
plt.title('Sales Analysis')
plt.xlabel('Months')
plt.ylabel("Amount in Lac")
plt.grid()
plt.show()
```

```python
months = np.arange(1,16)
sales = [253,645,267,879,798,776,534,327,334,562,677,637,576,584,364]
c = np.random.randint(10,100,15)
```
