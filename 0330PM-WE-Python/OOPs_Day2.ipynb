```python
# Single Inheritance , Multiple , Multilevel , Hybrid
```

```python
# Multilevel
class A:   #   Parent class
    def fun1(self):
        print("I am class A")
class B(A):    # Child class  (A child can access his parent's property)
    def fun2(self):
        print("I am class B")
class C(B):    # Child class  (A child can access his parent's property)
    def fun3(self):
        print("I am class C")
```

```python
# Multiple
class A:   #   Parent class
    def fun1(self):
        print("I am class A")
class B(A):    # Child class  (A child can access his parent's property)
    def fun2(self):
        print("I am class B")
class C(A):    # Child class  (A child can access his parent's property)
    def fun3(self):
        print("I am class C")
```

```python
# Hybrid
class A:   #   Parent class
    def fun1(self):
        print("I am class A")
class B(A):    # Child class  (A child can access his parent's property)
    def fun2(self):
        print("I am class B")
class C(A):    # Child class  (A child can access his parent's property)
    def fun3(self):
        print("I am class C")
class D(B,C):    # Child class  (A child can access his parent's property)
    def fun4(self):
        print("I am class D")
```

```python
obj = D()
```

```python
obj.fun1()
```

```python
obj.fun2()
```

```python
obj.fun3()
```

```python
obj.fun4()
```

```python
# Abstraction :- Data Hiding
```

```python
from abc import ABC , abstractmethod
class name(ABC):
    @abstractmethod
    def getDocument(self):
        pass
    def welcome(self):
        print("Welcome to My Portal")
```

```python
class name2(name):
    def getDocument(self):
        pass
    def myfun(self):
        print("This is child properties")
```

```python
obj = name2()
obj.myfun()
```

```python
obj.welcome()
```

```python
obj2 = name()
```
