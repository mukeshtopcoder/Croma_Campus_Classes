```python
import pandas as pd
import requests as r
from bs4 import BeautifulSoup
```

```python
url = "https://github.com/topics"
```

```python
r.get(url)
```

```python
content = r.get(url)
```

```python
content
```

```python
content.text[:500]
```

```python
soup = BeautifulSoup(content.text , "html.parser")
```

```python
len(soup)
```

```python
links = []
topics = []
desc = []
for val in soup.find_all('a',{'class':'no-underline flex-1 d-flex flex-column'}):
    links.append("https://github.com"+val['href'])
    topics.append(val.find('p').text)
    desc.append(val.find_all('p')[1].text.strip())
```

```python
len(topics) , len(desc) , len(links)
```

```python
data = {
    'Topics':topics,
    'Description':desc,
    'Links':links
}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("Github_Topics.csv")
```

```python
df
```

```python
users = []
repo = []
stars = []
links = []
```

```python
for url in df['Links']:
    content = r.get(url)
    soup = BeautifulSoup(content.text , 'html.parser')
    for val in soup.find_all('div',{'class':'d-flex flex-justify-between flex-items-start flex-wrap gap-2 my-3'}):
        users.append(val.find('h3').find('a').text)
        repo.append(val.find('h3').find_all('a')[1].text)
        stars.append(val.find('span',{'id':'repo-stars-counter-star'})['title'])
        links.append('https://github.com'+val.find('h3').find_all('a')[1]['href'])
```

```python
len(users) , len(repo) , len(stars) , len(links)
```

```python
newDf = pd.DataFrame(
    {
        'Users':users,
        "Repository":repo,
        'Rating':stars,
        'Links':links
    }
)
```

```python
newDf
```

```python
newDf.to_csv('All_Top_Repos.csv', index=False)
```
