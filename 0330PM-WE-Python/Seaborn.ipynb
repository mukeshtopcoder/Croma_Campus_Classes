```python
data = sns.load_dataset('tips')
```

```python
data.shape
```

```python
data.head(4)
```

```python
plt.figure( figsize=(6,3) )
sns.relplot(x="total_bill" , y='tip' , data=data)
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.relplot(x="total_bill" , y='tip' , data=data , hue='time')
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.relplot(x="total_bill" , y='tip' , data=data , hue='day')
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.relplot(x="total_bill" , y='tip' , data=data , hue='day' , col='time' , row='sex')
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.relplot(x="total_bill" , y='tip' , data=data , hue='day' , size='size')
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.relplot(x="total_bill" , y='tip' , data=data , kind="scatter" )
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.relplot(x="total_bill" , y='tip' , data=data , kind="line" )
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

```python
plt.figure( figsize=(6,3) )
markers = {'Lunch':'s' , 'Dinner':'X'}
sns.relplot(x="total_bill" , y='tip' , data=data , style='time' ,markers=markers)
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x = df.sepal_length , y = df.sepal_width)
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x = df.sepal_length , y = df.sepal_width , hue=df.species)
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x = df.sepal_length , y = df.sepal_width , hue=df.species , style=df.species)
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.show()
```
