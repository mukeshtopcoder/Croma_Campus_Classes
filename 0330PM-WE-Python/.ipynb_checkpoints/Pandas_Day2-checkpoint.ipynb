```python
# Object fill all missing value using MODE
# Numerical fill all missing value using MEAN
```

```python
# Filling missing value in Categorical/object/string
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(value=df[col].mode()[0]) 
    print(col," filled by the mean =>",df[col].mode()[0])
```

```python
# Filling missing value in Numerical/float64
for col in df.select_dtypes(include="float64").columns:
    df[col]=df[col].fillna(value=df[col].mean())
    print(col,"\t filled by mean/average value\t",round(df[col].mean(),2))
```

```python
import numpy as np
```

```python
# Repair the dataset
df.select_dtypes(include=np.number).head(3)
```

```python
df['Units Sold'].mean()
```

```python
df['new_Units'] = df['Gross Sales']/df['Sale Price']
```

```python
df.select_dtypes(include=np.number).head(3)
```

```python
for col in df.columns:
    if df[col].isnull().sum()>0:
        print(col,df[col].dtype,df[col].isnull().sum())
```

```python
# flinning missing value in Units Sold from columns new_Units
for i in range(len(df['Units Sold'])):
    if pd.isna(df['Units Sold'][i]):
        df['Units Sold'][i] = df['new_Units'][i]
```

```python
for col in df.columns:
    if df[col].isnull().sum()>0:
        print(col,df[col].dtype,df[col].isnull().sum())
```

```python
df= df.drop(columns=['new_Units'])
```

```python
for col in df.columns:
    if df[col].isnull().sum()>0:
        print(col,df[col].dtype,df[col].isnull().sum())
```
