```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
```

```python
df.shape
```

```python
df.head(3)
```

```python
plt.figure( figsize=(6,3) )
sns.barplot(x="class" , y="fare" , data = df)
plt.xlabel("CLASS")
plt.ylabel("FARE")
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.barplot(x="class" , y="fare" , data = df , hue="sex")
plt.xlabel("CLASS")
plt.ylabel("FARE")
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.distplot(df['fare'])
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.displot(df['fare'])
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.displot(df['fare'] , kde=True , color='green')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.histplot(df['fare'] )
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.histplot(df['fare'] , kde=True , color='green')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.histplot(df['fare'] , kde=True , color='green' , bins=20)
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.kdeplot(df['fare'] , color='green')
plt.show()
```

```python
df.head(3)
```

```python
plt.figure( figsize=(6,3) )
sns.kdeplot(x='fare' , color='green' , hue='class' ,  data=df , fill=True , alpha=0.6)
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.boxplot( y='fare' , data=df )
plt.show()
```

```python
plt.figure( figsize=(6,3) )
sns.boxplot( y='fare' , x='class' , data=df )
plt.show()
```
