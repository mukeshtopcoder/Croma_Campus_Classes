```python
data = {
    'EmpID':[101,102,103,104,105,106,107,108] , 
    'EmpName':['Raman','Shyam','Mohan',np.nan,'Ravi','Yogesh','Riya','Siya'],
    'EmpSalary':[23455,76555,np.nan,68765,64656,35556,np.nan,47645]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
print( df.isnull().sum().sum() )
```

```python
# Remove All Null Values
df.dropna()
```

```python
# replace missing value with a dummy value
newdf = df.fillna(value='Missing')
```

```python
newdf
```

```python
newdf.info()
```

```python
df
```

```python
newdf = df.copy()
```

```python
newdf
```

```python
newdf['EmpName'] = newdf['EmpName'].fillna('Missing')
```

```python
newdf
```

```python
newdf['EmpSalary'] = newdf['EmpSalary'].fillna( value = df['EmpSalary'].mean() )
```

```python
newdf
```
