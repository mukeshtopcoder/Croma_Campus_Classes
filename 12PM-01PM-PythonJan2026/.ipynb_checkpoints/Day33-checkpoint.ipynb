```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = {
    'Months':['Jan','Feb','Mar','April','May','Jun'],
    'Sales':[23455,56755,87335,47354,13144,86278]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.bar( df['Months'] , df['Sales'] )
```

```python
plt.figure( figsize=(6,3) )
plt.bar( df['Months'],df['Sales'] )
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.pie( df['Sales'] )
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.pie( df['Sales'] , labels=df['Months'])
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.pie( df['Sales'] , labels=df['Months'] , autopct="%0.2f%%" )
plt.show()
```

```python
plt.figure( figsize=(7,4) )
plt.pie( df['Sales'] , labels=df['Sales'] , autopct="%0.2f%%" )
plt.legend(df['Months'])
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot( df['Months'] , df['Sales'] )
plt.show()
```

```python
y = np.random.randint(1000,10000,30)
y
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y)
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , 'o')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.scatter(x,y)
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y,'o' , color='red')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , 'o')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , 'o:r')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , marker='o')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , marker='o' , mfc='red')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , marker='o' , mfc='red' , mec='black')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , marker='o' , mfc='red' , mec='black' , color='green')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , marker='o' , mfc='red' , mec='black' , color='green' , linestyle='dotted')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , marker='o' , mfc='red' , mec='black' , color='green' , linestyle='dashed')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , marker='o' , mfc='red' , mec='black' , color='green' , ls='-')
plt.show()
```

```python
plt.figure( figsize=(6,3) )
plt.plot(x,y , marker='o' , mfc='red' , mec='black' , color='green' , ls='-.')
plt.show()
```
