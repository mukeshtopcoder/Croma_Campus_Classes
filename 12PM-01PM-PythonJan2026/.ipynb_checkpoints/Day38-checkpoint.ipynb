```python
data = {
    'Month':['Jan','Feb','Mar','Apr','May'],
    'Sales':[24346,67565,36454,78976,24355]
}
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,3))
plt.bar( df['Month'] , df['Sales'] )
for i in range(len(df['Sales'])):
    plt.text( i , df['Sales'][i]+1000 ,df['Sales'][i] , ha='center'   )
plt.xlabel('Month Names')
plt.ylabel("Sales Rupee INR")
plt.title("Sales By Month")
plt.show()
```
