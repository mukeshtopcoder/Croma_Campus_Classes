### Data Analysis Project : Restaurent Tips Analysis

##### Objective
- Do males tips more than female?
- Does day affect tips?
- Does bigger group give larger tips?
- Does time (Lunch/Dinner) affect tips?

```python
# Dataset
import seaborn as sns
df = sns.load_dataset('tips')
df.head(3)
```

```python
# Importing Libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Gather information about dataset
df.info()
```

```python
# Dataset's Numerical Values Variance
df.describe()
```

```python
# Total Null Values in this dataset
print("Null Values :",df.isnull().sum().sum())
```

```python
# Columns in Dataset
df.columns
```

##### Total_Bill Distribution

```python
plt.figure( figsize=(6,3) )
sns.histplot( df['total_bill'] , kde=True )
plt.title("Total Bill Distribution")
plt.show()
```

##### Total_Bill Vs Tips

```python
plt.figure( figsize=(6,3) )
sns.scatterplot( x='total_bill' , y='tip' , data=df )
plt.title("Total Bill VS Tips")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

##### Tips By Gender

```python
plt.figure( figsize=(6,3) )
sns.boxplot( x='sex' , y='tip' , data=df )
plt.xlabel("Gender")
plt.ylabel("Tip")
plt.title("Tips by Gender")
plt.show()
```

##### Tips By Day

```python
plt.figure( figsize=(6,3) )
sns.boxplot( x='day' , y='tip' , data=df , palette='coolwarm' )
plt.xlabel("Day")
plt.ylabel("Tip")
plt.title("Tips by Day")
plt.show()
```

```python
##### Tips by Time
plt.figure( figsize=(6,3) )
sns.boxplot( x='time' , y='tip' , data=df , palette='viridis' )
plt.xlabel("Time")
plt.ylabel("Tip")
plt.title("Tips by Time")
plt.show()
```

##### PairPlot Analysis

```python
plt.figure( figsize=(6,3) )
sns.pairplot(df)
plt.show()
```

##### Correlation Heatmap Analysis

```python
corr = df.corr( numeric_only = True )
plt.figure( figsize=(6,3) )
sns.heatmap(corr , annot=True )
plt.show()
```

#### Insights
- Higher Total_Bill -> Higher_tips
- Dinner tips higher than Lunch
- Group Size Affect Total_Bill and Tips
- Weekends Tips are Higher
