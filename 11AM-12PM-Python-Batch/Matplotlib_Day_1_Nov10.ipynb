### SUBPLOT

```python
sales = np.array([23,56,89,56,73])
profit = np.array([57,36,78,45,68])
```

```python
plt.figure(figsize=(6,3))
plt.plot(sales)
plt.plot(profit)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.plot(sales , label='Sales')
plt.plot(profit , label='Profit')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.subplot(2,1,1)
plt.plot(profit)
plt.ylabel("Profit")
plt.subplot(2,1,2)
plt.plot(sales)
plt.ylabel("Sales")
plt.show()
```

```python
y1 = np.random.randint(100,500,20)
x1 = np.random.randint(1,10,20)
y2 = np.random.randint(100,500,20)
x2 = np.random.randint(1,10,20)
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x1,y1,label='Profit')
plt.scatter(x2,y2,label='sales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x1,y1,label='Profit',color='lightgreen')
plt.scatter(x2,y2,label='sales',color='hotpink')
plt.legend()
plt.show()
```

```python
x = np.random.randint(1,10,10)
y = np.random.randint(100,500,10)
colors=['green','red','black','orange','blue','skyblue','violet','pink','black','yellow']
```
