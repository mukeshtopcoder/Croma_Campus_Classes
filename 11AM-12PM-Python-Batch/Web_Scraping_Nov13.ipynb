```python
import requests as r
import pandas as pd
```

```python
url = 'https://github.com/topics'
```

```python
data = r.get(url)
```

```python
data
```

```python
data.content[:500]
```

```python
from bs4 import BeautifulSoup
```

```python
data = BeautifulSoup(data.content,'html.parser')
```

```python
len(data)
```

```python
data.find('a',class_='no-underline flex-1 d-flex flex-column')
```

```python
topics = []
desc = []
links = []
```

```python
extracted_data = data.find_all('a',class_='no-underline flex-1 d-flex flex-column')
```

```python
for ele in extracted_data:
    links.append('https://github.com'+ele['href'])
    topics.append(ele.find('p').text)
    desc.append((ele.find_all('p')[1].text).strip())
```

```python
len(links) , len(topics) , len(desc)
```

```python
dic = {
    'Topics':topics,
    'Description':desc,
    'Links':links
}
```

```python
df = pd.DataFrame(dic)
```

```python
df
```

```python
url = "https://github.com/topics/awesome"
```

```python
stars = []
users = []
repo = []
links = []
```

```python
for url in df['Links']:
    data = r.get(url)
    soup = BeautifulSoup(data.content,'html.parser')
    for ele in soup.find_all("div",class_='d-flex flex-justify-between flex-items-start flex-wrap gap-2 my-3'):
        users.append(ele.find('h3').find('a').text)
        repo.append(ele.find('h3').find_all('a')[1].text)
        links.append('https://github.com'+ele.find('h3').find_all('a')[1]['href'])
        stars.append(ele.find('span',{'id':'repo-stars-counter-star'})['title'])
```

```python
dic = {
    'Users':users,
    'Repository':repo,
    'Rating':stars,
    'Links':links
}
```

```python
df = pd.DataFrame(dic)
```
