```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set_theme(style='whitegrid',context='notebook')
```

```python
# Load dataset
df = sns.load_dataset("penguins")
print(df.shape)
df.head()
```

```python
# Cleaning
print(df.info())
print(df.isna().sum())
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Derived Columns (bill_ratio)
df['bill_ratio'] = df['bill_length_mm']/df['bill_depth_mm']
```

```python
df.head()
```

```python
# 
fig , axes = plt.subplots(1,3 , figsize=(15,4))
sns.histplot(df['bill_length_mm'] ,  kde=True , ax=axes[0])
axes[0].set_title('Bill Length Distribution')
sns.boxplot(x = df['species'] , y=df['bill_length_mm'] , ax=axes[1])
axes[1].set_title('Bill Length By Species')
sns.violinplot(x='species' , y='body_mass_g' , data=df , ax=axes[2])
axes[2].set_title('Body Mass By Species')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(data=df , x='flipper_length_mm' , y='body_mass_g' , hue='species')
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.scatterplot(data=df , x='flipper_length_mm' , y='body_mass_g' , hue='species', style='sex')
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.scatterplot(data=df , x='flipper_length_mm' , y='body_mass_g' , hue='species', style='sex')
sns.regplot(data=df , x = 'flipper_length_mm' , y= 'body_mass_g' , scatter=False , ci=50 , color='gray' )
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.scatterplot(data=df , x='flipper_length_mm' , y='body_mass_g' , hue='species', style='sex')
sns.regplot(data=df , x = 'flipper_length_mm' , y= 'body_mass_g' , scatter=False , ci=50 , color='gray' )
plt.title('Flipper Length VS Body Mass (With Regression)')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.scatterplot(data=df , x='flipper_length_mm' , y='body_mass_g' , hue='species', style='sex')
sns.regplot(data=df , x = 'flipper_length_mm' , y= 'body_mass_g' , scatter=False , ci=50 , color='gray' )
plt.title('Flipper Length VS Body Mass (With Regression)')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
# Annotate mean point
mean_x = df['flipper_length_mm'].mean()
mean_y = df['body_mass_g'].mean()
plt.scatter(mean_x,mean_y,s=100 , facecolors='none' , edgecolors='k', linewidths=1.5)
plt.annotate("Dataset Mean" , xy=(mean_x,mean_y) , xytext=(mean_x+1 , mean_y+100) , arrowprops=dict(arrowstyle='->' , lw=1))
plt.tight_layout()
plt.show()
```
