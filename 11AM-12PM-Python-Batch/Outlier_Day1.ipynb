```python
# IQR (Interquantile Range)
Q1 = df['EMP_Salary'].quantile(0.25)
Q3 = df['EMP_Salary'].quantile(0.75)
IQR_Q2 = Q3-Q1
```

```python
Q1 , Q3 , IQR_Q2
```

```python
min_value = Q1-IQR_Q2*1.5
max_value = Q3+IQR_Q2*1.5
```

```python
min_value , max_value
```

```python
newdf = df[df['EMP_Salary']<max_value]
```

```python
# Z_Score
```

```python
# z = (X-miu)/Sigma

# z = (data - mean)/std
```

```python
df['z_score'] = ( df['EMP_Salary'] - df['EMP_Salary'].mean() )/df['EMP_Salary'].std()
```

```python
df
```

```python
newdf = df[df['z_score']<2]
```
