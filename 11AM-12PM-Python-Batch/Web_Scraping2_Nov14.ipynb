```python
import pandas as pd
import requests as r
from bs4 import BeautifulSoup
```

```python
url = "https://www.boxofficemojo.com/year/"
```

```python
data = r.get(url)
```

```python
data
```

```python
data.content[:500]
```

```python
data = BeautifulSoup(data.content,'html.parser')
```

```python
year = []
totalGross = []
ly = []
release = []
average = []
topmovie = []
link = []
```

```python
for x in data.find_all('tr')[1:]:
    year.append(x.find_all('td')[0].text)
    totalGross.append(x.find_all('td')[1].text)
    ly.append(x.find_all('td')[2].text)
    release.append(x.find_all('td')[3].text)
    average.append(x.find_all('td')[4].text)
    topmovie.append(x.find_all('td')[5].text.strip())
    link.append('https://boxofficemojo.com/year/'+x.find_all('td')[0].text)
```

```python
len(year),len(totalGross),len(ly),len(release),len(average),len(topmovie),len(link)
```

```python
dataset = {
    'Year':year,
    'Total_Gross':totalGross,
    'LY':ly,
    'Total_Release':release,
    'Average_Gross':average,
    'Year_Top_Movie':topmovie,
    'Link':link
}
```

```python
df = pd.DataFrame(dataset)
```

```python
df.shape
```

```python
df[:5]
```

```python
movie = []
gross = []
theater = []
release_date = []
distributor = []
```

```python
for url in df['Link']:
    data = r.get(url)
    data = BeautifulSoup(data.content,'html.parser')
    for row in data.find_all('tr')[1:]:
        movie.append(row.find_all('td')[1].text.strip())
        gross.append(row.find_all('td')[5].text.strip())
        theater.append(row.find_all('td')[6].text.strip())
        release_date.append(row.find_all('td')[8].text.strip()+"  "+url[-4:])
        distributor.append(row.find_all('td')[9].text.strip())
```

```python
dataset = {
    'Movie_Name':movie,
    'Gross_Amount':gross,
    'Total_Theaters':theater,
    'Release_Date':release_date,
    'Distributor':distributor
}
```

```python
df = pd.DataFrame(dataset)
```
