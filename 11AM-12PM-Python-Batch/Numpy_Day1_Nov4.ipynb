```python
# array , arange , linspace
```

```python
arr = np.array([34,56,67,83])
print(arr)
```

```python
arr = np.arange(2,21,2)
print(arr)
```

```python
arr = np.linspace(0.5,5,10)
# linspace(start,stop,number_of_Values) # start and stop are included
print(arr)
```

```python
arr = np.linspace(0.5,5,25)
print(arr)
```

```python
# Dimensions
```

```python
# 0thD , 1stD , 2ndD , 3rdD etc
# Array can work on any dimension
```

```python
arr = np.array(46)
print(arr)
print(np.ndim(arr))
print(arr.shape)
```

```python
arr = np.array([23,45,67])
print(arr)
print(np.ndim(arr))
print(arr.shape)
```

```python
arr = np.array([[23,45,67],[56,78,90],[34,89,35]])
print(arr)
print(np.ndim(arr))
print(arr.shape)
```

```python
arr = np.array([[[32,45,67]]])
print(arr)
print(np.ndim(arr))
print(arr.shape)
```
