```python
seg = ['Government', 'Midmarket', 'Channel Partners', 'Enterprise','Small Business']
```

```python
seg
```

```python
# OHE One hot encoding
```

```python
for col in seg:
    df[col] = df['Segment'].map(lambda x:1 if x==col else 0)
```

```python
df.head()
```

```python
df = df.drop(columns=['Segment'])
```

```python
df['Discount Band'].unique()
```

```python
band = ['Low', 'Medium', 'High']
```

```python
df['Dis_Band'] = df['Discount Band'].map(lambda x: band.index(x))
```
