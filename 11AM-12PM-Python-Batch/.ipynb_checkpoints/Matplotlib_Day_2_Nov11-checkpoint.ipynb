```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales = np.array([27853,76243,58679,34778,28638,45868])
months = np.array(["Jan",'Feb','Mar','Apr','May','Jun'])
```

```python
plt.figure(figsize=(6,3))
plt.bar(months,sales)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(months,sales , color='red')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.barh(months,sales , color='red')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales , labels=months)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales , labels=sales)
plt.show()
```

```python
label = []
for i in range(0,len(months)):
    label.append(str(months[i])+" "+str(sales[i]))
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales , labels=months)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales , labels=label)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.pie(sales ,labels=months ,autopct="%.2f%%")
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Croma Campus\Downloads\Financial_Sample (2).xlsx")
```

```python
df.shape
```

```python
df = df.bfill()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
pbs = df.groupby(['Segment']).agg({'Profit':'sum'}).reset_index()
```

```python
pbs
```

```python
plt.figure(figsize=(6,3))
plt.bar(pbs['Segment'],pbs['Profit'],color='green')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.plot(pbs['Segment'],pbs['Profit'],marker='o')
plt.grid()
plt.show()
```

```python
pbp = df.groupby(['Product']).agg({'Profit':'sum'}).reset_index()
```

```python
plt.figure(figsize=(6,3))
plt.pie(pbp['Profit'] , autopct="%.2f%%" , labels=pbp['Product'])
plt.show()
```
