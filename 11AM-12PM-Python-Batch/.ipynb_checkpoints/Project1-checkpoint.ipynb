```python
def checkGroup(age):
    if age<30:
        return 'Youngesters'
    elif age<50:
        return 'Middle_Age'
    else:
        return 'Senior Citizen'
```

```python
df['Age_Group'] = df['Age'].map(checkGroup)
```

```python
df.head()
```

```python
pba = df.groupby(['Age']).agg({'Purchase Amount (USD)':'sum'}).reset_index()
```

```python
plt.figure(figsize=(7,2.5))
plt.plot(pba['Age'],pba['Purchase Amount (USD)'],marker='o',mfc='lightgreen',color='black')
plt.show()
```

```python
pba['Purchase Amount (USD)'].mean()
```

```python
pba = df.groupby(['Age_Group']).agg({'Purchase Amount (USD)':'sum'}).reset_index()
```

```python
pba
```

```python
plt.figure(figsize=(3,2.5))
fig , ax = plt.subplots()
ax.pie(pba['Purchase Amount (USD)'],labels=pba['Purchase Amount (USD)'])
circle = plt.Circle((0,0),0.60,fc='white')
plt.legend(pba['Age_Group'])
fig.gca().add_artist(circle)
plt.show()
```
