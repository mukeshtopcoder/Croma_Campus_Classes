```python
# x-> to_be_predict_for_x , y=predicted_value , m=slope , c = intercept
```

how to find m (slope)<br>
m = (n*(Exy) - Ex*Ey)/(n*E(x**2)-(Ex)**2)

```python
n = len(x)
```

```python
sum_xy = sum([x[i]*y[i] for i in range(n)])
sum_x = sum(x)
sum_y = sum(y)
sum_x2 = sum([i**2 for i in x])
```

```python
m = (n*sum_xy - sum_x*sum_y)/(n*sum_x2 - sum_x**2)
```

```python
m
```

c (intercept) <br>
c = (Ey-mEx)/n

```python
c = (sum_y-m*sum_x)/n
```

```python
c
```

```python
print("Slope = ",m)
print("Intercept = ",c)
```

```python
x = int(input("Enter Year/X : "))
y = m*x+c
print("Predicted Marks/Salary/Price : ",y)
```
