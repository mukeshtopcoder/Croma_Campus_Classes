```python
# Method to transform Votes from K,M to numeric value
def changeVotes(val):
    val = str(val)
    if val[-1]=='K':
        return int(float(val[0:len(val)-1])*1000)
    elif val[-1]=='M':
        return int(float(val[0:len(val)-1])*1000000)
    else:
        return int(val)
```

```python
changeVotes('297')
```

```python
df['Votes'] = df['Votes'].map(changeVotes)
```

```python
print("Average Votes for a Movie :",int(df['Votes'].mean()))
```

```python
# Method to change Duration
def changeDuration(val):
    val = val.split()
    if len(val)==2:
        return int(val[0][0:len(val[0])-1])*60 + int(val[1][0:len(val[1])-1])
    elif val[0][-1]=='h':
        return int(val[0][0:len(val[0])-1])*60
    elif val[0][-1]=='m':
        return int(val[0][0:len(val[0])-1])
    else:
        return changeDuration(df['Duration'].mode()[0])
```

```python
changeDuration('1h 23m')
```

```python
df['Duration'] = df['Duration'].map(changeDuration)
```

```python
df['Duration'].mean()
```

```python
print("Average Duration :",int(df['Duration'].mean()/60),'h',int(df['Duration'].mean())%60,'m')
```
