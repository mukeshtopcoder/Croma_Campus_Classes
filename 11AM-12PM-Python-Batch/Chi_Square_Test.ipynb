#### CHI-SQUARE TEST

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
dataset = {
    "Gender":['Male','Male','Female','Female'],
    "Purchased":["Yes",'No',"Yes",'No'],
    "Count":[20,30,40,10]
}
```

```python
df = pd.DataFrame(dataset)
```

```python
df
```

```python
contingency = pd.pivot_table(df,values="Count",
                            index="Gender",
                            columns="Purchased",
                            aggfunc="sum")
```

```python
contingency
```

```python
chi2 , p , dof , expected = chi2_contingency(contingency)
print("Chi-Square Value :",chi2)
print("P Value :",p)
print("Degree of Freedom :",dof),
print("Expected Value :",expected)
```

```python
alpha = 0.05
```

```python
if p < alpha :
    print("Strong Relationship")
else:
    print("No Relationship")
```

```python
# Null Hypothesis  (H0)        No Effect / No Relationship
# Alternative Hypothesis (H1)  Effect / Relationship   
```

```python
print("Chi2_Test Confidance : ",100-p)
```
