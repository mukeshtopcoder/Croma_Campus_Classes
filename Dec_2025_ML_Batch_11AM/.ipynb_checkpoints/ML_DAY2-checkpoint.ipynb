```python
plt.figure(figsize=(6,3))
sns.boxplot(df['Units Sold'])
plt.show()
```

```python
li = []
for col in df.select_dtypes(include=np.number).columns:
    if df[col].isnull().sum()>0:
        print(col)
        li.append(col)
```

```python
li
```

```python
plt.figure(figsize=(10,6))
for i in range(4):
    plt.subplot(2,2,i+1)
    sns.boxplot(df[li[i]])
plt.show()
```

```python
print("Col_Name \t Average\t STD")
for i in range(4):
    print(li[i],'\t',round(df[li[i]].mean(),2),'\t',round(df[li[i]].std(),2))
```
