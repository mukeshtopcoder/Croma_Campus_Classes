Milk , Butter <br>
Milk , Bread , Butter <br> 
Milk , Butter , Jam <br>
Milk , Jam  <br>
Milk , Bread , Butter , Jam <br>

{Milk,Bread} Antecedents -> Jam (Consequent)

Milk , Butter , Bread , Jam <br>
1      1        0       0   <br>
1      1        1       0   <br>
1      1        0       1   <br>
1      0        0       1   <br>
1      1        1       1   <br>

```python
data = [
['Milk' , 'Butter'],
['Milk' , 'Bread' , 'Butter'],
['Milk' , 'Butter' , 'Jam'],
['Milk' , 'Jam'],
['Milk' , 'Bread' , 'Butter' , 'Jam']
]
```

```python
data
```

```python
products = []
for sale in data:
    for product in sale:
        if product not in products:
            products.append(product)
products
```

```python
df = {}
for pro in products:
    li = []
    for sale in data:
        if pro in sale:
            li.append(1)
        else:
            li.append(0)
    df.update({pro:li})
```

```python
df
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(df)
```
