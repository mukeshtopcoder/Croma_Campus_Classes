```python
# Years
x = [1,2,3,4,5,6,7]
```

```python
# Salary
y = [30,35,40,45,50,55,60]
```

```python
# Linear Regression =>   y = mx+c
```

```python
import numpy as np
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
m = np.sum((x-x_mean)*(y-y_mean))/np.sum((x-x_mean)**2)
```

```python
m
```

```python
c = y_mean-m*x_mean
```

```python
c
```

```python
def pred(x_test):
    return m*x_test+c
```

```python
pred(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x,y)
plt.plot([8,9],[pred(8),pred(9)],ls='--',color='r')
plt.show()
```
