```python
# Ordinal Encoding
```

```python
object_col = df.select_dtypes(include="object").drop(columns=['furnishingstatus']).columns
```

```python
object_col
```

```python
for col in object_col:
    df[col] = df[col].map(lambda x:1 if x in ['yes','Yes','YES'] else 0)
```

```python
df.head()
```

```python
df.furnishingstatus.unique()
```

```python
def encodeFurnish(val):
    return list(df.furnishingstatus.unique()).index(val)
```

```python
df['furnishingstatus']=df['furnishingstatus'].map(encodeFurnish)
```
