```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('titanic')
```

```python
df = df.ffill()
```

```python
df.shape
```

```python
df.head(2)
```

```python
df.columns
```

```python
X = df[['age','fare','class']]
y = df['survived']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OrdinalEncoder
```

```python
X['class'].unique()
```

```python
sc = StandardScaler()
oe = OrdinalEncoder(categories=[['First','Second','Third']])
model = LogisticRegression()
```

```python
X[['age','fare']] = sc.fit_transform(X[['age','fare']])
```

```python
X.head(2)
```

```python
X['class'] = oe.fit_transform(X[['class']])
```

```python
X.head(2)
```

```python
X_train, X_test , y_train, y_test = train_test_split(X,y,random_state=11 , test_size=0.2)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.pipeline import Pipeline
```

```python
X = df[['age','fare']]
y = df['survived']
```

```python
X.head(2)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=11,test_size=0.20)
```

```python
pipeline = Pipeline([
    ('scaler',StandardScaler()),
    ('model',LogisticRegression())
])
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.compose import ColumnTransformer
```

```python
X = df[['age','fare','sex','embarked']]
y = df['survived']
```

```python
X.head(2)
```

```python
X.info()
```

```python
numerical_feature = X.select_dtypes(include=np.number).columns
numerical_feature
```

```python
categorical_feature = X.select_dtypes(include='object').columns
categorical_feature
```

```python
num_pipeline = Pipeline([
    ('missing',SimpleImputer(strategy='mean')),
    ('scaler',StandardScaler())
])
```

```python
cat_pipeline = Pipeline([
    ('missing',SimpleImputer(strategy='most_frequent')),
    ('encode',OrdinalEncoder())
])
```

```python
preprocessor = ColumnTransformer([
    ('num',num_pipeline,numerical_feature),
    ('cat',cat_pipeline,categorical_feature)
])
```

```python
pipeline = Pipeline([
    ('preprocessor',preprocessor),
    ('model',LogisticRegression())
])
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,random_state=11,test_size=0.2)
```

```python
X_train.head(2)
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'model__penalty': ['l1', 'l2'],
    'model__C': [0.01, 0.1, 1, 10, 100],
    'model__solver': ['liblinear', 'lbfgs'],
    'model__max_iter': [100, 200, 500]
}
```

```python
grid = GridSearchCV(
    pipeline,
    param_grid=param_grid , 
    cv = 5 , 
    scoring='accuracy',
    n_jobs=-1
)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
