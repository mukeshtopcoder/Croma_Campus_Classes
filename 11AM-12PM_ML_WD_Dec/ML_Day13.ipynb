### Unsupervised Learning 
#### Clustering

#### 1- Partition Based Clustering
- Number of K-Groups
- Fast and Widely Used
- Each point is related to a cluster
#### 2- Hierarchile Clustering
- No need to define K Groups
- Build a Tree of clusters (Dendogram)
- Easy to interpret
- Slow for large Datasets
#### 3- Density-Based Clustering
- Group points based on density
- Detects noise and outliers well
- Handles irregular shapes
- Struggle with varying densities

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_blobs( n_samples = 500 , random_state=42  )
```

```python
X[:4]
```

```python
scaled_X = StandardScaler().fit_transform(X)
```

```python
plt.figure(figsize=(6,3))
plt.scatter( scaled_X[:,0] , scaled_X[:,1] )
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.scatter( scaled_X[:,0] , scaled_X[:,1] , c=y)
plt.show()
```

```python
# Partition Based Clustering
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans
```

```python
kmeans.fit(scaled_X)
```

```python
labels = kmeans.fit_predict(scaled_X)
```

```python
plt.figure(figsize=(6,3))
plt.scatter( scaled_X[:,0] , scaled_X[:,1] , c=labels)
plt.show()
```

```python
# Hierarchile Clustering
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters = 4)
```

```python
labels = ac.fit_predict(scaled_X)
```

```python
plt.figure(figsize=(6,3))
plt.scatter( scaled_X[:,0] , scaled_X[:,1] , c=labels)
plt.show()
```
