```python
# Reinforcement Learning
```

```python
import numpy as np
import random as r
```

```python
player = 0
reward = 0
```

```python
table = []
while player != 4:
    step = r.randint(0,1)
    player = player+step
    if step == 1:
        reward+=10
    else:
        reward-=5
    table.append([step,reward])
    print("Player is on position",player)
    print("Player Reward is",reward)
else:
    print("Player is on position 4 now!")
```
