## Unsupervised Learning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans , DBSCAN , AgglomerativeClustering , SpectralClustering
from sklearn.decomposition import PCA
```

```python
data = datasets.load_iris()
X = data.data
y = data.target
```

```python
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X)
```

```python
X_pca[:3]
```

```python
sc = MinMaxScaler()
```

```python
X_scaled = sc.fit_transform(X_pca)
```

```python
X_scaled[:3]
```

```python
plt.figure(figsize=(6,3))
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=y  )
plt.show()
```

```python
kmeans = KMeans( n_clusters = 3 , random_state=11 )
k_cluster = kmeans.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=y  )
plt.subplot(1,2,2)
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=k_cluster )
plt.show()
```

```python
agg = AgglomerativeClustering(n_clusters=3)
agg_cluster = agg.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=y  )
plt.subplot(1,2,2)
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=agg_cluster )
plt.show()
```

```python
sp = SpectralClustering( n_clusters=3 )
sp_cluster = sp.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=y  )
plt.subplot(1,2,2)
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=sp_cluster )
plt.show()
```

```python
db = DBSCAN(eps=0.8 , min_samples=5)
db_cluster = db.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=y  )
plt.subplot(1,2,2)
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=db_cluster )
plt.show()
```

```python
db = DBSCAN(eps=0.8 , min_samples=5)
db_cluster = db.fit_predict(X_pca)
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=y  )
plt.subplot(1,2,2)
plt.scatter( X_scaled[:,0] ,X_scaled[:,1] , c=db_cluster )
plt.show()
```
